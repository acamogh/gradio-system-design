```python
import gradio as gr, base64, os, re, json
from groq import Groq
from google.colab import userdata

try:
    GROQ_API_KEY = userdata.get('GROQ_API_KEY')
except Exception:
    GROQ_API_KEY = "YOUR_GSK_KEY_HERE"

if not GROQ_API_KEY or GROQ_API_KEY == "YOUR_GSK_KEY_HERE":
    raise ValueError("⚠️ Setup Error: Insert your Groq token key inside the Colab Secrets panel.")

client = Groq(api_key=GROQ_API_KEY)
VISION_MODEL, TEXT_MODEL = "qwen/qwen3.6-27b", "openai/gpt-oss-120b"

DRIVE_FOLDER = "/content/drive/MyDrive"
MISTAKES_FILE = os.path.join(DRIVE_FOLDER, "system_design_mistakes.md")
HISTORY_FILE = os.path.join(DRIVE_FOLDER, "system_design_session.json")

if not os.path.exists(MISTAKES_FILE):
    with open(MISTAKES_FILE, "w", encoding="utf-8") as f:
        f.write("# 📑 Personal System Design Interview Error Log & Recap Guide\n\n")

SYSTEM_PROMPT = """You are an elite Technical Interview Panel. Singly guide candidate to BEST production architecture. Do not use Junior/Senior/Architect labels. Loop critique: brutally point out missing blocks/SPOF on uploaded screens or text. Demand Functional/Non-Functional metrics before drawing. Challenge every tool choice (Why Kafka/Redis?). Drill core areas: protocols, scale, storage, sharding, caching, observability. If explicit error made, prefix with: '=== MISTAKE ENCOUNTERED ===' followed by engineering fix details."""

def encode_image_to_base64(image_path):
    with open(image_path, "rb") as img_file: return base64.b64encode(img_file.read()).decode('utf-8')

def append_mistake_to_drive(mistake_text):
    try:
        with open(MISTAKES_FILE, "a", encoding="utf-8") as f:
            f.write(f"### ❌ Flagged Architecture Gap / Concept Drill\n{str(mistake_text).strip()}\n--- \n\n")
    except Exception as e: print(f"Log sync fault: {e}")

def load_history_from_drive():
    if not os.path.exists(HISTORY_FILE): return []
    try:
        with open(HISTORY_FILE, "r", encoding="utf-8") as f: raw_data = json.load(f)
        msg_list = []
        for item in raw_data:
            if isinstance(item, dict) and "role" in item and "content" in item:
                msg_list.append(item)
            elif isinstance(item, (list, tuple)) and len(item) >= 2:
                msg_list.append({"role": "user", "content": str(item[0])})
                msg_list.append({"role": "assistant", "content": str(item[1])})
        return msg_list
    except Exception: return []

def save_history_to_drive(history_data):
    try:
        compliant_data = []
        for turn in history_data:
            if isinstance(turn, dict) and "role" in turn and "content" in turn:
                compliant_data.append(turn)
            elif isinstance(turn, (list, tuple)) and len(turn) >= 2:
                compliant_data.append({"role": "user", "content": str(turn[0])})
                compliant_data.append({"role": "assistant", "content": str(turn[1])})
        with open(HISTORY_FILE, "w", encoding="utf-8") as f:
            json.dump(compliant_data, f, ensure_ascii=False, indent=2)
    except Exception as e: print(f"State synchronization fault: {e}")

def system_design_bot(message, history):
    if isinstance(message, str): message = {"text": message, "files": []}
    elif isinstance(message, dict) and "text" not in message: message = {"text": message.get("content", ""), "files": message.get("files", [])}

    groq_messages = [{"role": "system", "content": SYSTEM_PROMPT}]
    for turn in history:
        if isinstance(turn, dict):
            role, content = turn.get('role'), turn.get('content')
            if role == 'user' and content:
                text = content.get('text', '') if isinstance(content, dict) else content
                if isinstance(text, str) and text: groq_messages.append({"role": "user", "content": [{"type": "text", "text": text}]})
            elif role == 'assistant' and isinstance(content, str) and content:
                groq_messages.append({"role": "assistant", "content": content})
        elif isinstance(turn, (list, tuple)) and len(turn) >= 2:
            user_txt, bot_txt = turn[0], turn[1]
            if isinstance(user_txt, dict): user_txt = user_txt.get('text', '')
            if isinstance(bot_txt, dict): bot_txt = bot_txt.get('text', '')
            if user_txt and isinstance(user_txt, str): groq_messages.append({"role": "user", "content": [{"type": "text", "text": user_txt}]})
            if bot_txt and isinstance(bot_txt, str): groq_messages.append({"role": "assistant", "content": bot_txt})

    current_content, has_image = [], False
    if message.get("text"): current_content.append({"type": "text", "text": message["text"]})
    if message.get("files"):
        for file_info in message["files"]:
            try:
                img_path = file_info.get("path") if isinstance(file_info, dict) else file_info
                if img_path and os.path.exists(img_path):
                    current_content.append({"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{encode_image_to_base64(img_path)}"} })
                    has_image = True
            except Exception as e: current_content.append({"type": "text", "text": f"[Upload error: {str(e)}]"})

    if not current_content: current_content.append({"type": "text", "text": "Let's kick off the interview loop."})
    groq_messages.append({"role": "user", "content": current_content})

    try:
        completion = client.chat.completions.create(model=VISION_MODEL if has_image else TEXT_MODEL, messages=groq_messages, temperature=0.6, max_completion_tokens=2500, stream=True)
        text_reply = ""
        for chunk in completion:
            if hasattr(chunk, 'choices') and chunk.choices:
                choice = chunk.choices[0]
                delta = getattr(choice, 'delta', None)
                chunk_text = getattr(delta, 'content', '') if delta else choice.get('delta', {}).get('content', '')
                if chunk_text: text_reply += chunk_text; yield text_reply

        if "=== MISTAKE ENCOUNTERED ===" in text_reply:
            parts = text_reply.split("=== MISTAKE ENCOUNTERED ===")
            if len(parts) > 1: append_mistake_to_drive(" ".join(parts[1:]))

        updated_history = history + [{"role": "user", "content": message.get("text", "")}, {"role": "assistant", "content": text_reply}]
        save_history_to_drive(updated_history)
    except Exception as e: yield f"⚠️ Interface Error: {str(e)}"

def delete_and_reset_history():
    if os.path.exists(HISTORY_FILE): os.remove(HISTORY_FILE)
    if os.path.exists(MISTAKES_FILE):
        with open(MISTAKES_FILE, "w", encoding="utf-8") as f: f.write("# 📑 Personal System Design Interview Error Log & Recap Guide\n\n")
    return [], "🛡️ Session logs reset complete."

def load_recap_log():
    status_header = "## 📈 Gaps & Refinement Summary\n\n"
    if not os.path.exists(MISTAKES_FILE): return status_header + "No flaws logged yet!"
    with open(MISTAKES_FILE, "r", encoding="utf-8") as f: content = f.read()
    return status_header + content

fullscreen_css = """
@import url('https://googleapis.com');
* { font-family: 'Google Sans', 'Product Sans', 'Open Sans', sans-serif !important; }
.gradio-container { max-width: 100% !important; padding: 10px 2% !important; }
.tabs, .tabitem, .tabitem > div, .chatbot, [data-testid="chatbot"], .chatbot > .wrap, .chatbot .bubble-wrap, .chatbot .panel {
    height: 80vh !important; min-height: 80vh !important; max-height: 80vh !important;
}
.chatbot .bubble-wrap { overflow-y: auto !important; }
"""

with gr.Blocks(theme=gr.themes.Soft(), css=fullscreen_css) as demo:
    gr.Markdown("### 🏢 Continuous System Design Optimization Terminal")
    with gr.Tab("🔴 Architecture Review Loop"):
        chat = gr.ChatInterface(fn=system_design_bot, type="messages", multimodal=True, textbox=gr.Textbox(placeholder="Click '+' to attach diagrams or type 'Let's begin'..."))
    with gr.Tab("📊 Architecture Recap Log & Controls"):
        recap_box = gr.Markdown(value="Click the button below to parse your historical refinement logs.")
        with gr.Row():
            refresh_btn = gr.Button("🔄 Generate Recap & View System Flaws", variant="primary")
            reset_btn = gr.Button("🗑️ Clear Session & Start Fresh", variant="stop")
        refresh_btn.click(fn=load_recap_log, outputs=recap_box)
        reset_btn.click(fn=delete_and_reset_history, outputs=[chat, recap_box])
    demo.load(fn=load_history_from_drive, outputs=chat.chatbot)

demo.queue().launch(inline=True, share=True, debug=True)
```
