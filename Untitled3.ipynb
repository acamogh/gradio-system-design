```python
import gradio as gr, base64, os, re
from groq import Groq
from google.colab import userdata

try:
    GROQ_API_KEY = userdata.get('GROQ_API_KEY')
except Exception:
    GROQ_API_KEY = "YOUR_GSK_KEY_HERE"

if not GROQ_API_KEY or GROQ_API_KEY == "YOUR_GSK_KEY_HERE":
    raise ValueError("⚠️ Setup Error: Insert your Groq token key inside the Colab Secrets panel.")

client = Groq(api_key=GROQ_API_KEY)
VISION_MODEL, TEXT_MODEL = "qwen/qwen3.6-27b", "openai/gpt-oss-120b"
DRIVE_FOLDER = "/content/drive/MyDrive"
MISTAKES_FILE = os.path.join(DRIVE_FOLDER, "system_design_mistakes.md")

if not os.path.exists(MISTAKES_FILE):
    with open(MISTAKES_FILE, "w", encoding="utf-8") as f:
        f.write("# 📑 Personal System Design Interview Error Log & Recap Guide\n\n")

SYSTEM_PROMPT = """You are an unyielding, elite Technical Interview Panel. Your singular goal is to guide the candidate to build the absolute BEST, most optimal, production-grade system architecture possible. Do not label the candidate as Junior, Senior, or Architect. Focus entirely on iterative perfection.

CRITICAL ENGAGEMENT PROTOCOL:
1. REQUIREMENTS SELECTION: Present 3 major system design prompts (e.g., Global Video Streaming, Real-time Ride-Hailing, Distributed Rate Limiter). Force them to select one and define the exact Functional & Non-Functional metrics (QPS, Latency, SLA, Storage) before drawing anything.
2. CONTINUOUS LOOP CRITIQUE: Once they share their high-level blocks or upload an architecture screenshot, evaluate it brutally. Point out missing components, scalability bottlenecks, data consistency flaws, or single points of failure (SPOF). Keep prompting them about what they missed UNTIL the design handles extreme failure modes cleanly.
3. THE "WHY" COMPONENT CHALLENGE: Challenge them to defend every tool they add (e.g., Kafka, Redis, Cassandra). Ask: 'Why this specific tool over alternative components?', 'What are the architectural trade-offs?'.
4. DEEP CORE TESTING: Continuously cross-examine their evolving layout across the 9 major system design areas: Networking Protocols, Traffic Sizing, Databases, Sharding/Fault Tolerance, Distributed Consistency (CAP/PACELC), Cache Invalidation, Event Broker Guarantees, Reliability, and Observability.

MISTAKE LOGGING:
If the candidate makes a significant conceptual error, prefix your critique with this exact tag: '=== MISTAKE ENCOUNTERED ===' followed by the details of the gap."""

def encode_image_to_base64(image_path):
    with open(image_path, "rb") as img_file: return base64.b64encode(img_file.read()).decode('utf-8')

def append_mistake_to_drive(mistake_text):
    try:
        with open(MISTAKES_FILE, "a", encoding="utf-8") as f:
            f.write(f"### ❌ Flagged Architecture Gap / Concept Drill\n{str(mistake_text).strip()}\n--- \n\n")
    except Exception as e: print(f"Error saving log slice: {e}")

def system_design_bot(message, history):
    # 1. Normalize current user query payload formats
    if isinstance(message, str): message = {"text": message, "files": []}
    elif isinstance(message, dict) and "text" not in message: message = {"text": message.get("content", ""), "files": message.get("files", [])}

    groq_messages = [{"role": "system", "content": SYSTEM_PROMPT}]

    # 2. FIXED: Fully Bulletproof History Structure Parser
    for turn in history:
        if isinstance(turn, dict):
            # Handler for list-of-dictionary structures
            role = turn.get('role')
            content = turn.get('content')
            if role == 'user' and content:
                text = content.get('text', '') if isinstance(content, dict) else content
                if isinstance(text, str) and text:
                    groq_messages.append({"role": "user", "content": [{"type": "text", "text": text}]})
            elif role == 'assistant' and content:
                if isinstance(content, str) and content:
                    groq_messages.append({"role": "assistant", "content": content})

        elif isinstance(turn, (list, tuple)) and len(turn) >= 2:
            # Handler for standard alternating sub-array blocks: [user_text, bot_text]
            user_text = turn[0]
            bot_text = turn[1]

            # Unpack dictionary wrappers if Gradio formats multi-modal items inside the past list arrays
            if isinstance(user_text, dict):
                user_text = user_text.get('text', '')
            if isinstance(bot_text, dict):
                bot_text = bot_text.get('text', '')

            if user_text and isinstance(user_text, str):
                groq_messages.append({"role": "user", "content": [{"type": "text", "text": user_text}]})
            if bot_text and isinstance(bot_text, str):
                groq_messages.append({"role": "assistant", "content": bot_text})

    # 3. Assemble current incoming elements
    current_content, has_image = [], False
    if message.get("text"): current_content.append({"type": "text", "text": message["text"]})
    if message.get("files"):
        for file_info in message["files"]:
            try:
                img_path = file_info.get("path") if isinstance(file_info, dict) else file_info
                if img_path and os.path.exists(img_path):
                    current_content.append({"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{encode_image_to_base64(img_path)}"} })
                    has_image = True
            except Exception as e: current_content.append({"type": "text", "text": f"[Upload error: {str(e)}]"})

    if not current_content: current_content.append({"type": "text", "text": "Let's kick off the optimization interview."})
    groq_messages.append({"role": "user", "content": current_content})

    try:
        completion = client.chat.completions.create(model=VISION_MODEL if has_image else TEXT_MODEL, messages=groq_messages, temperature=0.6, max_completion_tokens=2500, stream=True)
        text_reply = ""
        for chunk in completion:
            if hasattr(chunk, 'choices') and chunk.choices:
                choice = chunk.choices[0]
                delta = getattr(choice, 'delta', None)
                chunk_text = getattr(delta, 'content', '') if delta else choice.get('delta', {}).get('content', '')
                if chunk_text:
                    text_reply += chunk_text
                    yield text_reply

        if "=== MISTAKE ENCOUNTERED ===" in text_reply:
            parts = text_reply.split("=== MISTAKE ENCOUNTERED ===")
            if len(parts) > 1: append_mistake_to_drive(" ".join(parts[1:]))
    except Exception as e: yield f"⚠️ Interface Error: {str(e)}"

def load_recap_log():
    status_header = "## 📈 System Design Gaps & Refinement Summary\n*(Below are the architecture design limits we flagged during your session)*\n\n"
    if not os.path.exists(MISTAKES_FILE): return status_header + "No flaws logged yet! Keep designing to test system extremes."
    with open(MISTAKES_FILE, "r", encoding="utf-8") as f: content = f.read()
    return status_header + content

fullscreen_css = """
@import url('https://googleapis.com');
* { font-family: 'Google Sans', 'Product Sans', 'Open Sans', sans-serif !important; }

.gradio-container { max-width: 100% !important; padding: 10px 2% !important; }

.tabs,
.tabitem,
.tabitem > div,
.chatbot,
[data-testid="chatbot"],
.chatbot > .wrap,
.chatbot .bubble-wrap,
.chatbot .panel {
    height: 80vh !important;
    min-height: 80vh !important;
    max-height: 80vh !important;
}

.chatbot .bubble-wrap {
    overflow-y: auto !important;
}
"""

with gr.Blocks(theme=gr.themes.Soft(), css=fullscreen_css) as demo:
    gr.Markdown("### 🏢 Continuous System Design Optimization Terminal")

    with gr.Tab("🔴 Architecture Review Loop"):
        chat = gr.ChatInterface(
            fn=system_design_bot,
            type="messages",
            multimodal=True,
            textbox=gr.Textbox(placeholder="Click the clip/plus icon to attach diagrams or type 'Let's begin'...")
        )
    with gr.Tab("📊 Architecture Recap Log"):
        recap_box = gr.Markdown(value="Click the button below to parse your historical refinement logs.")
        refresh_btn = gr.Button("🔄 Generate Recap & View System Flaws", variant="primary")
        refresh_btn.click(fn=load_recap_log, outputs=recap_box)

demo.queue().launch(inline=True, share=True, debug=True)
```
