````python
import gradio as gr, base64, os, json, random
from groq import Groq
from google.colab import userdata
import re

# ==========================================
# 1. SETUP CREDENTIALS & POOL
# ==========================================
try:
    GROQ_KEY = userdata.get('GROQ_API_KEY')
except Exception:
    GROQ_KEY = "YOUR_GSK_KEY_HERE"

if not GROQ_KEY or GROQ_KEY == "YOUR_GSK_KEY_HERE":
    raise ValueError("⚠️ Set GROQ_API_KEY in Secrets.")

client = Groq(api_key=GROQ_KEY)

TEXT_MDL = "openai/gpt-oss-120b"

FOLDER = "/content/drive/MyDrive"
LOG_FILE = os.path.join(FOLDER, "system_design_mistakes.md")

if not os.path.exists(LOG_FILE):
    try:
        os.makedirs(FOLDER, exist_ok=True)
        with open(LOG_FILE, "w", encoding="utf-8") as f:
            f.write("# 📑 System Design Error Log\n\n")
    except Exception:
        LOG_FILE = "system_design_mistakes.md"
        with open(LOG_FILE, "w", encoding="utf-8") as f:
            f.write("# 📑 System Design Error Log (Local Fallback)\n\n")

SYSTEM_POOL = [
    "Real-Time Ride-Hailing Service (like Uber/Lyft)",
    "Global Video Streaming Architecture (like YouTube/Netflix)",
    "Distributed Key-Value Cache Layer (like Redis Clusters)",
    "Massive Scale Chat Messaging System (like WhatsApp/Telegram)",
    "Infinite Scroll Video Feed Platform (like TikTok/Instagram Reels)",
    "High-Throughput Distributed Rate Limiter Service",
    "Real-time Collaboration Workspace Backend (like Google Docs/Figma)",
    "Global E-Commerce Flash Sale & Inventory System (like Amazon)",
    "Distributed Web Crawler and Search Indexer (like Google Crawler)",
    "Location-Based Nearby Business Discovery Engine (like Yelp/Maps)"
]

def handle_shuffle_and_clear():
    new_system = random.choice(SYSTEM_POOL)
    return (
        new_system,
        "",
        "",
        "*Feedback cleared...*",
        "",
        "*Math feedback cleared...*",
        "### System changed!\nSelect a teaching phase below and click 'Teach Me This Phase' to break it down."
    )

def log_err(txt):
    try:
        with open(LOG_FILE, "a", encoding="utf-8") as f:
            f.write(f"### ❌ Flagged Gap\n{str(txt).strip()}\n\n")
    except Exception: pass

# ==========================================
# 2. INTERVIEW CORE LOGIC (MARKDOWN-SAFE OUTPUTS)
# ==========================================
def review_stage1(name, functional, non_functional):
    if not name or not functional or not non_functional:
        return "⚠️ Provide Name, Functional and Non-Functional items."
    prompt = (
        f"You are a FAANG Panel testing requirements scale.\n"
        f"Target System: {name}\n"
        f"Candidate's Functional Requirements:\n{functional}\n\n"
        f"Candidate's Non-Functional Requirements:\n{non_functional}\n\n"
        f"INSTRUCTIONS:\n"
        f"1. Check functional and non-functional items strictly.\n"
        f"2. Add what is missing (SLA boundaries, scale targets, CAP stance).\n"
        f"3. Provide clean, baseline corrected criteria arranged in a structured Markdown table.\n"
        f"4. If gaps exist, start text with token: '=== MISTAKE ENCOUNTERED ==='."
    )
    try:
        completion = client.chat.completions.create(
            model=TEXT_MDL,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3,
            max_completion_tokens=1500
        )
        res = completion.choices[0].message.content
        if "=== MISTAKE ENCOUNTERED ===" in res:
            log_err(f"Stage 1 on {name}:\n{res}")
        return res
    except Exception as e: return f"⚠️ Error: {str(e)}"

def review_stage2(name, user_input):
    if not name or not user_input:
        return "⚠️ Provide Name and Calculations input."
    prompt = (
        f"You are a FAANG Panel testing estimations.\n"
        f"Target System: {name}\n"
        f"Candidate input:\n{user_input}\n\n"
        f"INSTRUCTIONS:\n"
        f"1. Audit math, splits (QPS), and storage limits.\n"
        f"2. Show formulas and correct calculations formatted cleanly using Markdown breakdown tables.\n"
        f"3. If math fails, start text with token: '=== MISTAKE ENCOUNTERED ==='."
    )
    try:
        completion = client.chat.completions.create(
            model=TEXT_MDL,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3,
            max_completion_tokens=1500
        )
        res = completion.choices[0].message.content
        if "=== MISTAKE ENCOUNTERED ===" in res:
            log_err(f"Stage 2 on {name}:\n{res}")
        return res
    except Exception as e: return f"⚠️ Error: {str(e)}"

# ==========================================
# TEACHER-DRIVEN PHASED ARCHITECTURE LOGIC (ADVANCED TECH ICONS)
# ==========================================
def generate_phased_design(name, selected_phase):
    if not name:
        return "⚠️ Please enter a Target System Name first.", ""

    prompt = (
f"You are a patient System Design mentor teaching a fresher or junior engineer.\n"
f"System: '{name}'\n"
f"Current Teaching Phase: {selected_phase}\n\n"
f"IMPORTANT: This is part of a progressive step-by-step series.\n\n"
f"INSTRUCTIONS (Follow strictly):\n"
f"1. Create a clean Mermaid diagram code block utilizing advanced system shape layout standards:\n"
f"   - DATABASE NODES: All databases MUST use the native Mermaid cylinder shape notation. Syntax: DB_ID[(\"Database Name\")]\n"
f"   - KAFKA / EVENT BUS NODES: Must use standard bracket quoting alongside FontAwesome message icon tags embedded inside. Syntax: Kafka_ID[\"fa:fa-comments Kafka / Event Bus\"]\n"
f"   - CLIENTS/SERVERS: Use standard FontAwesome icons where possible (e.g., Laptop icon: `fa:fa-laptop Client`, Server stack: `fa:fa-server Application Server`).\n"
f"   - CRITICAL QUOTING RULE: Every text label inside a node's container symbols MUST be wrapped in explicit double quotes.\n"
f"   - Example Format combinations:\n"
f"        UserNode[\"fa:fa-user Client App\"]\n"
f"        KafkaNode[\"fa:fa-share-alt Kafka Stream EventBus\"]\n"
f"        MainDB[(\"fa:fa-database Primary SQL DB\")]\n"
f"2. After the diagram, give 3-4 simple bullet points or structured table comparison explaining the current layer updates vs legacy structures.\n"
f"Output format: Start with ```mermaid then the diagram, then the markdown lesson content."
    )

    try:
        completion = client.chat.completions.create(
            model=TEXT_MDL,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.2,
            max_completion_tokens=1500
        )
        raw_text = completion.choices[0].message.content

        mermaid_code = ""
        if "```mermaid" in raw_text:
            parts = raw_text.split("```mermaid")
            if len(parts) > 1:
                mermaid_code = parts[1].split("```")[0].strip()

        explanation_markdown = raw_text.split("```")[-1].strip()

        if not mermaid_code:
            mermaid_code = "graph TD\nA[\"fa:fa-laptop Client\"] --> B[\"fa:fa-server API Gateway\"]"

        markdown_payload = f"## 🏫 Mentor Mode: {selected_phase}\n\n```mermaid\n{mermaid_code}\n```\n\n---\n\n### 🎓 Phase Breakdown & Interview Strategy\n{explanation_markdown}"
        return markdown_payload, mermaid_code
    except Exception as e:
        return f"❌ Teacher module hit a bump: {str(e)}", ""

def read_log():
    if not os.path.exists(LOG_FILE): return "No logs found."
    with open(LOG_FILE, "r", encoding="utf-8") as f: return f.read()

# ==========================================
# 3. INTERFACE BUILDER & CSS (TABLE + WRAP FIX)
# ==========================================
ui_css = """
/* Google Sans Typography setup without breaking markdown tables */
.gradio-container, .gradio-container h1, .gradio-container h2, .gradio-container h3,
.gradio-container p, .gradio-container label, .gradio-container span, .gradio-container button {
    font-family: 'Google Sans', '-apple-system', BlinkMacSystemFont, sans-serif !important;
}

/* Maintain proper code layout */
.out-box pre, .out-box code, .gr-textbox textarea {
    font-family: 'Fira Code', 'Courier New', monospace !important;
}

/* Robust Markdown Container Box */
.out-box {
    border: 1px solid #cbd5e1 !important;
    border-radius: 8px !important;
    padding: 16px !important;
    background-color: #f8fafc !important;
    max-height: 600px !important;
    overflow-y: auto !important;
    overflow-x: auto !important;
    width: 100% !important;
}

/* Force standard text wrapping layout across elements inside the output container */
.out-box p, .out-box li, .out-box pre, .out-box code {
    white-space: pre-wrap !important;
    word-break: break-word !important;
}

/* Bulletproof Markdown Table Containment - Keeps tables fully formed but scrollable horizontally if too wide */
.out-box table {
    display: block !important;
    width: 100% !important;
    max-width: 100% !important;
    overflow-x: auto !important;
    white-space: nowrap !important;
    border-collapse: collapse !important;
    margin: 15px 0 !important;
}
.out-box th, .out-box td {
    border: 1px solid #cbd5e1 !important;
    padding: 8px 12px !important;
    text-align: left !important;
    white-space: normal !important; /* Forces the text *inside* cells to wrap nicely */
}
.out-box th {
    background-color: #f1f5f9 !important;
    font-weight: bold !important;
}
"""

with gr.Blocks(theme=gr.themes.Soft(), css=ui_css) as demo:
    gr.Markdown("# 🏢 Modular FAANG System Design Workspace")

    with gr.Row():
        sys_box = gr.Textbox(
            label="🎯 Target System Name",
            value="Real-Time Ride-Hailing Service (like Uber/Lyft)",
            placeholder="Type your custom system here...",
            interactive=True,
            scale=4
        )
        shuffle_btn = gr.Button("🎲 Shuffle Next System Prompt", variant="secondary", scale=1)

    with gr.Tab("📋 Stage 1: Requirements"):
        gr.Markdown("### Step 1: Define Scope Criteria")
        with gr.Row():
            in1_func = gr.Textbox(label="Functional Requirements (Actions)", lines=4, placeholder="e.g., Users book rides, drivers accept...")
            in1_nonfunc = gr.Textbox(label="Non-Functional Requirements (SLA/CAP)", lines=4, placeholder="e.g., Latency under 200ms, eventual consistency fine...")
        btn1 = gr.Button("🔍 Submit Stage 1 Review", variant="primary")
        out1 = gr.Markdown("*Feedback will load here...*", elem_classes=["out-box"])
        btn1.click(review_stage1, [sys_box, in1_func, in1_nonfunc], out1)

    with gr.Tab("📊 Stage 2: Math Calculations"):
        gr.Markdown("### Step 2: Capacity Sizing Estimates")
        in2 = gr.Textbox(label="Calculations (QPS, DAU, Storage)", lines=5)
        btn2 = gr.Button("⚡ Audit Estimations", variant="primary")
        out2 = gr.Markdown("*Math feedback will load here...*", elem_classes=["out-box"])
        btn2.click(review_stage2, [sys_box, in2], out2)

    with gr.Tab("🏫 Stage 3: Step-by-Step Architecture Classroom"):
        gr.Markdown("### Step 3: Learn the Architecture Incrementally")

        with gr.Row():
            with gr.Column(scale=1, min_width=280):
                gr.Markdown("#### 🛠️ Lesson Selection Panel")
                phase_selector = gr.Dropdown(
                    label="Select Teaching Phase to Isolate",
                    choices=[
                        "Phase 0: High-Level Overview & Core Components",
                        "Phase 1: Client Side & API Gateway / Edge Layer",
                        "Phase 2: Authentication, Authorization & Rate Limiting",
                        "Phase 3: Load Balancers & Stateless Application Servers",
                        "Phase 4: Message Queues & Asynchronous Processing",
                        "Phase 5: Core Business Services & Microservices Design",
                        "Phase 6: Caching Strategy (Client + Server + CDN)",
                        "Phase 7: Database Selection, Sharding & Replication",
                        "Phase 8: Data Consistency, CAP Theorem & Trade-offs",
                        "Phase 9: Storage & File Handling (S3-like, Blob Storage)",
                        "Phase 10: Search & Indexing (Elasticsearch style)",
                        "Phase 11: Monitoring, Logging & Observability",
                        "Phase 12: Scaling, Deployment & Reliability Patterns"
                    ],
                    value="Phase 0: High-Level Overview & Core Components",
                    interactive=True
                )
                teacher_btn = gr.Button("💡 Teach Me This Phase", variant="primary")

                gr.Markdown("---")
                mermaid_copy_box = gr.Textbox(
                    label="📋 Raw Mermaid Code Blueprint",
                    lines=8,
                    show_copy_button=True,
                    interactive=False
                )

            with gr.Column(scale=2, min_width=320):
                gr.Markdown("#### 🗺️ Guided Architecture Blueprint & Lesson Plan")
                out_solution = gr.Markdown("Select a phase on the left and click 'Teach Me This Phase' to compile the layout cleanly.", elem_classes=["out-box"])

        teacher_btn.click(
            generate_phased_design,
            [sys_box, phase_selector],
            [out_solution, mermaid_copy_box]
        )

    with gr.Tab("📝 Drive Error Log Tracker"):
        out4 = gr.Markdown("Click Sync to read.", elem_classes=["out-box"])
        btn4 = gr.Button("🔄 Sync Mistakes File", variant="secondary")
        btn4.click(read_log, None, out4)

    shuffle_btn.click(
        fn=handle_shuffle_and_clear,
        inputs=None,
        outputs=[sys_box, in1_func, in1_nonfunc, out1, in2, out2, out_solution]
    )

if __name__ == "__main__":
    demo.queue().launch(inline=True, share=True, debug=True)
````
